```python
%matplotlib inline
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot.
fig = plt.figure()
ax = fig.add_subplot()
# Change the line color to green, market to diamond, line width to 2 points.
ax.plot(x_axis, y_axis, color="green", marker = "D", linewidth=2,)
# Add a legend for the city of Boston.
ax.legend("Boston")
# Add a title and axes labels.
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.set_title("PyBer Fare by Month")
# Add a y-axis limit.
ax.set_ylim(0,45)
# Add grid lines.
ax.grid()
```

```python
# Create the plot with scatter and color the points red
plt.scatter(y_axis, x_axis, color="red")
# Invert the y axis
plt.gca().invert_yaxis()
# Create labels for the x and y axes.
plt.xlabel("Fare($)")
plt.ylabel("Date")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend("Chicago")
# Add a limit for the x-axis
plt.xlim(0,50)
```

```python
# Create a bubble chart
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Create larger bubbles
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
# Plot the scatter plot
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Format for list comprehension
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Create scatterplot with object-oriented interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# create bubble chart using object-oriented interface
fig, ax = plt.subplots()
ax.scatter(
           # Switch the axis
           y_axis, x_axis, 
           # Change the color to blue
           color="skyblue", 
           # Change the size of the markers to 5 times each data point
           s=[i * 5 for i in y_axis], 
           # Make the color 20% transparent
           alpha=.8,
           # Add a black edge color to the circles
           edgecolors="black",
           # Make the linewidth 2
           linewidths=2)
# Add a legend for the city of Boston.
ax.legend("Boston")
# Add a title.
ax.set_title("PyBer Fare by Month")
# Add a limit for the x-axis data.
ax.set_xlim(0,50)
# Invert the y-axis so that January is at the top.
ax.invert_yaxis()
# Create labels for x and y axis
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
# Increase the figure size to 8x8
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    # Explode the two highest percentage months
    explode=explode_values,
    colors=colors,
    # Add a shadow
    shadow=True,
    # Start angle so Jan is at the top
    startangle=90,
    labels=x_axis,
    # Reverse month order
    counterclock=False,
    # Add a percentage to one decimal place to each wedge of the pie
    autopct='%.1f%%')
plt.show()
```

```python
# Choose wedges to explode
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
# Chaneg colours
colors=["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
# Increase the figure size to 8x8
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(y_axis,labels=x_axis,
    explode=explode_values,
    colors=colors,
    # Add a shadow
    shadow=True,
    # Start angle so Jan is at the top
    startangle=90,
    # Reverse month order
    counterclock=False,
    # Add a percentage to one decimal place to each wedge of the pie
    autopct='%.1f%%')
plt.show
```
